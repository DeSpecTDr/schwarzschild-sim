```python
import numpy as np
from scipy.integrate import quad,romberg
import matplotlib.pyplot as plt

GM = 3.986004e+14
rp = 6500000
vp = 10000
# GM = 10.0
# rp = 1.0
# vp = 1.0
L = rp * vp
E0 = vp * vp / 2 - GM / rp
a = 1/(2/rp - vp*vp/GM)
e = 1 - rp/a
ra = a*(1+e)

def orbit(r):
    return L / (2*E0*r**4 + 2*GM*r**3 - L**2*r**2) ** 0.5

#res = quad(orbit, rp, ra)
r = np.linspace(rp, ra, 500)
o = np.array([quad(orbit, rp, ri)[0] if ri != rp else 0 for ri in r])
plt.polar(o, r / 6500000)
```
